**8. Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.**

```python
def sales_percentage(regionSales, globalSales):
    '''
    Calculate percentage of regionalSales in globalSales
    
    args:
        regionSales(float)
        globalSales(float)
        
    return:
        percentage(float): percentage of regional sales in global sales
    '''
    try:
        return (regionSales / globalSales) * 100
    except ZeroDivisionError:
        return 0
    
# calcluate percentage of Global Sales from North America
video_games_df['NA_Global_Sales_Percent'] = sales_percentage(video_games_df['NA_Sales'], video_games_df['Global_Sales'])
video_games_df.head()
```

**10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.**

```python
# Initial attempt for Mean User Score
print(f"Mean User Score: {video_games_df['User_Score'].mean()}")
```

```python
# View top 5 rows of User_Score
video_games_df['User_Score'].head()
```

```python
# convert non-numeric values to NaN
video_games_df['User_Score'] = pd.to_numeric(video_games_df['User_Score'], errors = 'coerce')
video_games_df['User_Score'].head()
```

```python
# Convert NaN values to the median User_Score
video_games_df['User_Score'] = video_games_df['User_Score'].fillna(video_games_df['User_Score'].median())
video_games_df['User_Score'].head()
```
